```

# =============================================================================
# 1. IMPORT LIBRARIES
# =============================================================================
import pandas as pd
import numpy as np
from pathlib import Path
from IPython.display import display

print("Libraries imported successfully.")

# =============================================================================
# 2. LOAD DATA FILES
# =============================================================================
data_dir = Path("../data")
assumptions_path = data_dir / "model_assumptions.csv"
archetypes_path = data_dir / "simulated_archetype_data.xlsx"

try:
    assumptions_df = pd.read_csv(assumptions_path)
    cement_df = pd.read_excel(archetypes_path, sheet_name='Cement_Sector')
    petrochem_df = pd.read_excel(archetypes_path, sheet_name='Petrochemical_Sector')
    utilities_df = pd.read_excel(archetypes_path, sheet_name='Utilities_Sector')
    print("✅ Data files loaded successfully.")
except FileNotFoundError as e:
    print(f"❌ File not found: {e}")

# =============================================================================
# 3. PREVIEW DATA
# =============================================================================
for name, df in {
    "Model Assumptions": assumptions_df,
    "Cement Sector": cement_df,
    "Petrochemical Sector": petrochem_df,
    "Utilities Sector": utilities_df
}.items():
    print(f"\n--- {name} ---")
    display(df.head())
```
